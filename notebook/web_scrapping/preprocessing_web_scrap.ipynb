# Preprocessing Web Scrapping data (elo_distribution_data.csv)

```python
import pandas as pd
```

```python
df_data = pd.read_csv("../../data/raw/elo_distribution_brut_data.csv")
df_data
```

```python
# Suppression de la premier colonne du df_data qui ne sert pas à grand choses
# Il y a les paliers I, II, III et IV de chaque palier, on a pas besoin
df_data = df_data.drop(df_data.columns[0], axis=1)
df_data
```

```python
# On veut supprimer toute les lignes ou il y a des Nan
df_data = df_data.dropna()
print("Après suppression des élement NaN")
df_data
```

```python
# Changer les noms de chaque palier, enlever les "I" apres chaque nom de palier
df_data = df_data.replace("Diamond I", "Diamond")
df_data = df_data.replace("Emerald I", "Emerald")
df_data = df_data.replace("Platinum I", "Platinum")
df_data = df_data.replace("Gold I", "Gold")
df_data = df_data.replace("Silver I", "Silver")
df_data = df_data.replace("Bronze I", "Bronze")
df_data = df_data.replace("Iron I", "Iron")
df_data
```

```python
# Suppression de la deuxieme colonne qui est inutile
df_data = df_data.drop(df_data.columns[1], axis=1)
df_data
```

```python
df = df_data.rename(columns={df_data.columns[0]: "Rank", df_data.columns[1]: "Tier %"})
df
```

```python
new_row = {'Rank': 'Challenger', 'Tier %': '0.00015%'}

# Créer un DataFrame avec la nouvelle ligne
new_row_df = pd.DataFrame([new_row])

# Ajouter la nouvelle ligne au début du DataFrame existant
df = pd.concat([new_row_df, df], ignore_index=True)

df
```

```python
df.to_csv('../../data/processed/clear_elo_distribution_data.csv', index=False)
```
